## Deleting Unnecessary Columns

```python
df.drop(['Series Name','Country Name'],axis=1,inplace=True)
```

## Top 10 Countries with Total Population

```python
# Filter data for total population
total_population_data = df[df['Series Code'] == 'SP.POP.TOTL']

# Sort data Based on the total population for 2020
total_population_sorted = total_population_data.sort_values(by="2020", ascending=False)

# Get the Top 10 Countries with the highest total population for 2020
total_top_ten_countries = total_population_sorted.head(10)
print("Top ten countries of total population\n")
print(total_top_ten_countries[['Country Code']])
```

## Bar Plot

## Top Ten Countries of Total Population in year 2019 and 2014

```python
# Create the bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2019", y="Country Code", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2019)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

# Create the bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2014", y="Country Code", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2014)",fontsize=10)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

## Top 10 Countries with highest male population

```python
# Filter data for Male population
male_population_data = df[df['Series Code'] == 'SP.POP.TOTL.MA.IN']

# Sort data Based on the male population for 2017
male_population_sorted = male_population_data.sort_values(by="2017", ascending=False)

# Get the Top 10 Countries with the highest male population for 2017
male_top_ten_countries = male_population_sorted.head(10)
print("Top ten countries of male population")
print(male_top_ten_countries[['Country Code']])
```

## Top 10 Countries with highest Female population in 2020

```python
# Filter data for Female population
female_population_data = df[df['Series Code'] == 'SP.POP.TOTL.FE.IN']

# Sort data Based on the Female population for 2017
female_population_sorted = female_population_data.sort_values(by="2020", ascending=False)

# Get the Top 10 Countries with the highest Female population for 2020
female_top_ten_countries = female_population_sorted.head(10)
print("Top ten countries of female population")
print(female_top_ten_countries[['Country Code']])
```

## Top 10 Countries with highest Male and Female population in 2020

```python
# Create the bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2020", y="Country Code", data=male_top_ten_countries, palette="viridis")
plt.title("Top Ten Countries of Male Population (2020)",fontsize=10)
plt.xlabel("Male Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

# Create the bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2020", y="Country Code", data=female_top_ten_countries, palette="viridis")
plt.title("Top Ten Countries of Female Population (2020)",fontsize=10)
plt.xlabel("Female Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

## Stacked Bar Plot

## Top 10 Countries With Male And Female Population

```python
# merge male and female population data on 'Country Code'
merge_data=pd.merge(male_population_data,female_population_data,on="Country Code",suffixes=("_male","_female"))
```

```python
# merged data
#Calculate the total population of each country (male + female)
merge_data["Total population"] = merge_data["2022_male"] + merge_data ["2022_female"]
```

```python
merge_data.head()
```

```python
# sort data based on total population in descending order
sorted_data = merge_data.sort_values(by="Total population", ascending=False)
```

## Select Top 10 Countries with the highest Population

```python
top_10_countries = sorted_data.head(10)
```

```python
# create the stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code", y="2022_female",data=top_10_countries,color="hotpink",label="female Population")
```

```python
# create the stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code", y="2022_female",data=top_10_countries,color="forestgreen",label="female Population")
sns.barplot(x="Country Code", y="2022_male",data=top_10_countries,bottom=top_10_countries["2022_female"],color="orangered",label="male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```

## Bottom 10 Countries with Male and Female Population (2020)

```python
# select Bottom 10 countries with male and female population
bottom_10_countries = sorted_data.tail(10)
```

```python
# create the stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code", y="2022_female",data=bottom_10_countries,color="hotpink",label="female Population")
sns.barplot(x="Country Code", y="2022_male",data=bottom_10_countries,bottom=bottom_10_countries["2022_female"],color="cyan",label="male Population")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```
